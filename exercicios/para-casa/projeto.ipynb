```python
conteudo = csv.reader(file)
```

```python
connection = sqlite3.connect ("_banco_dados_saude-sono.db")
```

```python
cursor = connection.cursor()
```

```python
cursor.execute('''
    CREATE TABLE IF NOT EXISTS qualidade_sono( 
        Person_ID INTEGER PRIMARY KEY,
        Gender VARCHAR,
        Age INTEGER,
        Occupation VARCHAR,
        Sleep_Duration REAL,
        Quality_of_Sleep TEXT,
        Physical_Activity_Level TEXT,
        Stress_Level TEXT,
        BMI_Category TEXT,
        Blood_Pressure TEXT,
        Heart_Rate INTEGER,
        Daily_Steps INTEGER,
        Sleep_Disorder TEXT
    )
''')

```

```python

inserir_conteudo = "INSERT INTO  qualidade_sono(Person_ID,Gender, Age, Occupation, Sleep_Duration, Quality_of_Sleep, Physical_Activity_Level, Stress_Level, BMI_Category, Blood_Pressure, Heart_Rate, Daily_Steps, Sleep_Disorder) VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?)"


```

```python
cursor.executemany(inserir_conteudo, conteudo)
```
